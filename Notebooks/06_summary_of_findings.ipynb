**Summary for Big Mountain Resort**
---
**Problem:**

Following the installation of a new chair lift that increased the distribution of visitors across the mountain, Big Mountain Resort was presented with the problem of offsetting the additional operating costs. With operating costs increasing by $1,540,000 due to the new chair lift, the costs needed to be offset through additional revenue by raising ticket prices or cost-cutting measures. Additionally, with average yearly visitors totaling 350,000, how much could ticket prices be increased without having a significant impact on attendance if increasing revenue was determined to be the solution?

**Recommendations:**

After wrangling and cleaning the available data on ski resorts throughout the United States, their various features, and pricing, it was revealed that increasing ticket prices to be the most viable solution. Upon determining an increase ticket prices was the most feasible of the two solutions, statistical modeling was required to identify the cost point that they could be raised to by developing an understanding of the impact various resort features have upon pricing dynamics.
To model ticket pricing, adult weekend pricing as the base of our model given the availability of data across resorts in comparison to the Adult Weekday price, which was missing significant values.

After training and testing the model, it was determined that Big Mountain's current ticket price of \\$81.00 for an adult weekend ticket can be raised to \\$95.87. While this would make it the most expensive ticket price for all Montana ski resorts, modeling indicated this would not impact visitor attendance and subsequently not lead to a negative impact on revenue.

To implement this increase, additional changes are needed to the resort, primarily the addition of new features. Based on our random forest model, an increase in ticket prices can only be justified by adding an additional run, installing an additional chair lift, and increasing the vertical drop by 150 feet. These changes should yield an additional \\$3,474,638 in revenue, which is more than double the \\$1,540,000 needed to offset the added operating costs from the new chair lift.
